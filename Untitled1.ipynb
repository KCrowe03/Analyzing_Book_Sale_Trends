```python
import sqlite3
import pandas as pd

# Load the first dataset
df1 = pd.read_csv(r"C:\Users\Kriwe\OneDrive\Documents\CodeKY Projects\CodeKY DA\Capstone\Datasets\kindle_data-v2.csv")

# Display the first few rows to understand its structure
df1.head()
```

```python
# Load the second dataset
df2 = pd.read_csv(r"C:\Users\Kriwe\OneDrive\Documents\CodeKY Projects\CodeKY DA\Capstone\Datasets\book1-100k.csv")

# Display the first few rows to understand its structure
df2.head()
```

```python
conn = sqlite3.connect('books_db.sqlite')
df1.to_sql('amazon_books', conn, index=False, if_exists='replace') 
df2.to_sql('goodreads_books', conn, index=False, if_exists='replace')
```

```python
print(df1.columns)
print(df2.columns)
```

```python
# Drop columns that aren't needed
df1.drop(columns=['imgUrl', 'productURL'], inplace=True)
df2.drop(columns=['RatingDist1', 'ISBN'], inplace=True)
```

```python
# print(df1.columns)
```

```python
# Add a 'source' column to distinguish the datasets
df1['source'] = 'Amazon'
df2['source'] = 'GoodReads'
```

```python
print(df1.head())
print(df2.head())
```

```python
# Convert 'publishedDate' to a datetime format, handling errors
df1['publishedDate'] = pd.to_datetime(df1['publishedDate'], errors='coerce')

# Extracting the year from 'publishedDate' and storing it in 'PublishYear'
df1['PublishYear'] = df1['publishedDate'].dt.year

# Display the first few rows to verify
print(df1[['publishedDate', 'PublishYear']].head())
```

```python
print(df1.isnull().sum())
print(df2.isnull().sum())
```

```python
df1['title'].head(15) 
df2['Name'].head(15)  
```

```python
import re

# Clean up df1's titles by removing anything in parentheses
df1['cleaned_title'] = df1['title'].apply(lambda x: re.sub(r'\(.*\)', '', x).strip())

# Clean up df2's titles removing anything in parentheses
df2['cleaned_name'] = df2['Name'].apply(lambda x: re.sub(r'\(.*\)', '', x).strip())

# Check the first 15 cleaned titles for both datasets
print(df1['cleaned_title'].head(15))
print(df2['cleaned_name'].head(15))
```

```python
merged_df = pd.merge(df1, df2, left_on='cleaned_title', right_on='cleaned_name', how='inner')
print(merged_df.head())
```

```python
# Check for missing data
print(merged_df.isnull().sum())

# Check for duplicates
duplicates = merged_df.duplicated()
print(f'Duplicates: {duplicates.sum()}')
```

```python
# Filtering the dataset to only include books that are part of the Kindle Unlimited program
kindle_unlimited_books = merged_df[merged_df['isKindleUnlimited'] == True]

# Filter by a specific genre or price range
high_priced_books = merged_df[merged_df['price'] > 20]
```
